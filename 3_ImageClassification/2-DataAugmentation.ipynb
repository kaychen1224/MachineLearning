# ImageDataGenerator

```
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,                               
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator()


training_set = train_datagen.flow_from_directory('/content/drive/My Drive/Dataset/ImageClassification/CatDog/Train',
                                                 target_size = (128, 128),
                                                 batch_size = 32,
                                                 class_mode = 'binary')

validation_set = test_datagen.flow_from_directory('/content/drive/My Drive/Dataset/ImageClassification/CatDog/Validation',
                                            target_size = (128, 128),
                                            batch_size = 32,
                                            class_mode = 'binary')
```
