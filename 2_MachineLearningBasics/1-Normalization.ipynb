# Normalization

```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

square = [29,20,18,14,44,25,70]
distance =  [4,5,1,2,3,2,8]


square_min = min(square)
square_max = max(square)
square_denominator = square_max - square_min

for index in range(0,len(square)):
    square[index] = (square[index]-square_min) / square_denominator

print(square)


distance_min = min(distance)
distance_max = max(distance)
distance_denominator = distance_max - distance_min

for index in range(0,len(distance)):
    distance[index] = (distance[index]-distance_min) / distance_denominator

print(distance)
```

# Standardization

```
from sklearn import preprocessing

data = [ [29,20,18,14,44,25,70], [4,5,1,2,3,2,8]]
data_std = preprocessing.scale(data, axis=1, with_mean=True, with_std=True)
print(data_std)

sum = 0.0
for index in range(0,7):
  sum = sum + data_std[0][index]

print(sum)
```

**Standardization練習**



```
from sklearn import preprocessing

data = [ [29,20,18,14,44,25,70], [4,5,1,2,3,2,8], [1200,800,750,960,850,450,980]]
data_std = preprocessing.scale(data, axis=1, with_mean=True, with_std=True, copy=True)
print(data_std)

sum = 0.0
for index in range(0,7):
  sum = sum + data_std[2][index]

print(sum)
```
