# LinearRegression

```
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt


# x = (店面面績, 車站距離)
X = np.array([[50, 5], [40, 1], [35, 2], [10, 1], [30, 3], [60, 12], [50, 10], [20, 3]])
#print(X.shape)
#print(X[0])

# y = 售價
y = np.array([555, 520, 371, 150, 246, 560, 520, 211])
#print(y.shape)

plt.title('Square feet vs Sales')
plt.scatter(X[:, 0], y)
plt.show()

plt.title('KM vs Sales')
plt.scatter(X[:, 1], y)
plt.show()

# 建立線性迴歸函數
lm = LinearRegression()
# 找出最適合的線性迴歸函數
lm.fit(X, y)
# 模型績效, mean square error 
mse = np.mean((lm.predict(X) - y) ** 2)
# 印出模型績效
print('mse',mse) # mean square error


# 新蛋糕店資料
new_store_1 = np.array([
    [50, 2]
])
new_store_1_sales = lm.predict(new_store_1)

# 預測新蛋糕店的單月銷量
print(new_store_1_sales)
```

```
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt

# x = (店面面績, 車站距離)
#X = np.array([[30, 15], [40, 1], [25, 10], [25, 8], [20, 2], [35, 3]])
X = np.array([[30], [40], [25], [25], [20], [35]])


# y = 售價
y = np.array([40, 200, 55, 70, 150, 180])

#plt.title('KM vs Sales')
#plt.scatter(X[:, 1], y)
#plt.show()

plt.title('Square feet vs Sales')
plt.scatter(X[:, 0], y)
plt.show()
```

```
# 建立線性迴歸函數
lm = LinearRegression()

# 找出最適合的線性迴歸函數
lm.fit(X, y)


# 模型績效, mean square error 
mse = np.mean((lm.predict(X) - y) ** 2)

# 印出模型績效
print(mse) # mean square error
```

```
# 新蛋糕店資料
new_store_1 = np.array([
    [10]
])
new_store_1_sales = lm.predict(new_store_1)

# 預測新蛋糕店的單月銷量
print(new_store_1_sales)
```
