# 迴圈

```
list1 = [1,2,3,4,5]
for index in range(0,4):
    list1[index] = list1[index]+1
print(list1)

list1 = [1,2,3,4,5]
index = 0
while index < 5:
    list1[index] = list1[index]+1
    index = index + 1
print(list1)
```

# 函數

```
def multiply(input1, input2):
    return (input1*input2)*(input1*input2)

print(multiply(2,3))
```

函數練習

```
def mean(input_list):
  sum = 0
  for index in range(0,len(input_list)):
    sum = sum + input_list[index]
  return sum/len(input_list)

list1 = [1,2,3,4]
print(mean(list1))
```

# 類別

```
class Car:
    def __init__(self, length, width, height, weight):
        self.length = length
        self.width = width
        self.height = height
        self.weight = weight

    def print_all(self):
        print('Car')
        print('length',self.length)
        print('width',self.width)
        print('height',self.height)
        print('weight',self.weight)

car = Car(10,20,30,40)
car.print_all()
```

類別練習

```
class Computer:
    def __init__(self, cpu_model, cpu_price, ram_model, ram_price):
        self.cpu_model = cpu_model
        self.cpu_price = cpu_price
        self.ram_model = ram_model
        self.ram_price = ram_price

    def cost(self):
        return self.cpu_price + self.ram_price
    

pc1 = Computer('cpu1',100,'ram1',222)
print(pc1.cost())
```
